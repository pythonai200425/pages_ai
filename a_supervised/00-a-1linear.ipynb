```python
import math

x = np.arange(0,9)

def f1(x):
    return 2*x -0.5
    
y1 = f1(x)

X = np.array([1,2,3,4]) #Features
Y = np.array([2,2,7,8]) #Labels

count = 0
for i in range len(X):
    plt.scatter(X[i],Y[i]) 
    
#sqr_errors = [0,4,1,0]
error = []
for i in range (len(X)):
    error_i = math.pow(Y[i] - f1(X[i]),2) #חישוב הפרש בין התצפית לבין הקו (המספר 2 זהו מספר החזקה)
    error.append(error_i)
error_mean = sum(error)/len(error)

print(error_mean)

plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x,y1, label = "y =  2*x")

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```

```python
import math

x = np.arange(0,9)

def f1(x):
    return 2.3*x -1
    
y1 = f1(x)

X = np.array([1,2,3,4]) #Features
Y = np.array([2,2,7,8]) #Labels

count = 0
for i in range (len(X)):
    plt.scatter(X[i],Y[i]) 
    
#sqr_errors = [0,4,1,0]
error = []
for i in range (len(X)):
    error_i = math.pow(Y[i] - f1(X[i]),2) #חישוב הפרש בין התצפית לבין הקו (המספר 2 זהו מספר החזקה)
    error.append(error_i)
error_mean = sum(error)/len(error)

print(error_mean)

plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x,y1, label = "y =  2.3x + 1")

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Our data
X = np.array([1, 2, 3, 4]).reshape(-1, 1)  
Y = np.array([2,2,7,8])  

# Create regression model
model = LinearRegression()
model.fit(X, Y)

# Print results
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")
# Calculate equation
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")
```
