```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Our data
X = np.array([2, 3, 5, 7, 9, 12, 16, 20, 25, 30]).reshape(-1, 1)
Y = np.array([95, 85, 70, 65, 60, 55, 50, 53, 58, 70])

'''
min_mse = None
a_min = None
b_min = None
c_min = None
for a in np.arange(-2, 2, 0.1):
    for b in np.arange(-10, 10, 0.1):
        for c in np.arange(-50, 105, 1):
            y_pred = get_y(a, b, c, np.array([2, 3, 5, 7, 9, 12, 16, 20, 25, 30]))
            cur_mse = mse(Y, y_pred)
            if min_mse == None or cur_mse < min_mse:
                min_mse = cur_mse
                a_min, b_min, c_min = a, b, c
                
'''         

# y = a*x**2  + b*x + c
# a = ? b = ? c = ?
# X = np.array([2, 3, 5, 7, 9, 12, 16, 20, 25, 30]).reshape(-1, 1)

# 11111111111111111111111111111111111111
# X = np.array([2, 3, 5, 7, 9, 12, 16, 20, 25, 30]).reshape(-1, 1)
# XÂ² 

poly = PolynomialFeatures(degree=2, include_bias=True)  # prepare
X_poly_3_column = poly.fit_transform(X)  # do

model = LinearRegression()
model.fit(X_poly_3_column, Y)

intercept = model.intercept_
coefficients = model.coef_

print(f"Intercept (Î²â) c: {intercept:.2f}")
print(f"Coefficient for x (Î²â)  b: {coefficients[1]:.2f}")
print(f"Coefficient for xÂ² (Î²â) a: {coefficients[2]:.2f}")
a = coefficients[2]
b = coefficients[1]
c = intercept
print(f"y = {a:.2f}xÂ² + {b:.2f}x + {c:.2f}")

# get r2 + adjr2
r2 = r2_score(Y, model.predict(X_poly_3_column))
print('r2', r2)

plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Observation points')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# find the y of the parabula using the formula 
y_graph = get_para(a, b, c, x)

# another way to find the y of the parabula and draw it
x_parabula = np.arange(0, 35)
x_parabula_reshape = x_parabula.reshape(-1,1)  # not always necessary 
x_parabula_split_3 = poly.fit_transform(x_parabula_reshape)
y_parabula = model.predict(x_parabula_split_3)

plt.plot(x_parabula, y_parabula, color='red', linewidth=2, label=f'{a:.2f}xÂ² + {b:.2f}x + {c:.2f}')

plt.legend()
plt.xlim(0, 40)
plt.ylim(00, 150)
plt.show()
```
